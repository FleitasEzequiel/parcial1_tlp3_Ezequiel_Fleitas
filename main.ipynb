```python
df["nombre"] = df["nombre"].str.capitalize()
df["apellido"] = df["apellido"].str.capitalize()
df["voto"] = df["voto"].str.upper()

df.dropna(inplace=True)
df.replace("SÍ","SI",inplace=True)
df.drop(labels=(df[(df.voto != "SI") & (df.voto != "NO")].index), inplace=True)
df.dni = df.dni.astype("int32")
```
